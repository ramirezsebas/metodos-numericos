```python
def euler(expr_xy, xi, xf, x0, y0, n, decimales=6):
    h = (xf - xi) / n
    
    def fxy(x, y):
        fxy = eval(expr_xy)
        return fxy
    
    X = [x0 + i*h for i in range(n+1)]
    Y = [0]*(n+1)
    Y[0] = y0
    
    print("\nSOLUCIÓN:")
    
    for k in range(1, n+1):
        print("\nIteración #%d\n" % k)
        
        Y[k] = round(Y[k-1] + h * fxy(X[k-1], Y[k-1]), decimales)
        print("y%d = %.{0}f + %.{0}f * f(%.{0}f, %.{0}f) = %.{0}f".format(decimales) % (k, Y[k-1], h, X[k-1], Y[k-1], Y[k]))
        
    print("\n\nI\t\txn\t\tyn")
    print("--------------------------------------------")
    print("0\t\t%.{0}f\t%.{0}f".format(decimales) % (X[0], Y[0]))
    print("--------------------------------------------")
    for k in range(1, n+1):
        print("%d\t\t%.{0}f\t%.{0}f".format(decimales) % (k, X[k], Y[k]))
        print("--------------------------------------------")
    
    print("\ny(%.{0}f) = %.{0}f".format(decimales) % (X[n], Y[n]))
    
expr_xy = input("Ingrese al función de (x, y): ")
print("\nCondiciones iniciales")
x0 = float(input("x0 = "))
y0 = float(input("y0 = "))
print("\nIntervalo")
xi = float(input("xi = "))
xf = float(input("xf = "))
print("\nIteraciones")
n = int(input("n = "))
euler(expr_xy, xi, xf, x0, y0, n)
```
