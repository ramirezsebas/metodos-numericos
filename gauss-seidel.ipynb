```python
from pprint import pprint
from numpy import array, zeros, diag, diagflat, dot, concatenate

def print_ecuaciones(A, b, decimales=6):
    AUG = concatenate((A, b),axis=1)

    for i in range(len(AUG)):
        print("x%d^(k) = " % (i+1), end='')
        for j in range(len(AUG[0])):
            if i == j or AUG[i][j] == 0:
                print("0", end='') if j == 0 else print(" + 0", end='')
            else:
                if j == 0:
                    print("%.{}fx%d".format(decimales) % (-AUG[i][j]/AUG[i][i], j), end='')
                else:
                    numerador = AUG[i][j] if j == len(AUG[0])-1 else -AUG[i][j]
                    
                    print(" + ", end='') if numerador/AUG[i][i] > 0 else print(" - ", end='')

                    print("%.{}fx%d".format(decimales) % (abs(numerador/AUG[i][i]), j), end='')
                print("^(k)", end='') if j < i else print("^(k-1)", end='')
        print("")

def gauss_seidel(A, b, N, decimales=6, x=None):
    
    if x is None:
        x = zeros(len(A[0]))
    
    print("\n\nSOLUCIÓN\n")
    print("Ecuaciones para la k-ésima iteración:\n")
    print_ecuaciones(A, b, decimales)
    
    print("\nIteraciones\n")
    b = b.flatten()
        
    for k in range(N):
        for i in range(len(x)):
            suma = 0
            for j in range(len(x)):
                if i != j:
                    suma += A[i][j] * x[j]
            x[i] = (-suma + b[i]) / A[i][i]
        print("iteracion %d ->\t" % (k+1), end='')
        for j in range(len(x)):
            print("%.{}f\t\t".format(decimales) % x[j], end='')
        print()
        

n = int(input("Cantidad de incógnitas: "))

A = zeros((n,n))
print('\nCoeficientes de la matriz A:')
for i in range(n):
    for j in range(n):
        A[i][j] = float(input( 'A['+str(i)+']['+ str(j)+']='))
            
b = zeros((n,1))
print('\nCoeficientes de la matriz b:')
for i in range(n):
    b[i][0] = float(input( 'b['+str(i)+']['+ str(j)+']='))
            

dec = int(input("\nCantidad de decimales: "))
iteraciones = int(input("\nCantidad de iteraciones: "))

# A = array([[10, -1, 2, 0],
#            [-1, 11, -1, 3],
#            [2, -1, 10, -1],
#            [0, 3, -1, 8]])

# b = array([[6], [25], [-11], [15]])

gauss_seidel(A, b, 5, 4)
# b = array([[6], [25], [-11], [15]])
# print_ecuaciones(A, b, 6)
```
